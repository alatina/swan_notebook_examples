```octave
% N = number of particles
% Q = tune
% R = transfer matrix
% X = gaussian bunch

% create a simple periodic lattice
Q = 0.2516;
phi = 2*pi*Q;
R = [ cos(phi) sin(phi) ; -sin(phi) cos(phi) ];
```

```octave
% create a bunch with N = 10000 particles
N = 10000;
X = 0.2 * randn(2,N);
```

```octave
% CASE 1: simulate 1000 turns just with linear optics
for i=1:1000
  X = R * X;
end
```

```octave
% plot
plot(X(1,:), X(2,:), '.');
axis([ -1 1 -1 1 ]);
xlabel('x [mm]');
ylabel('x'' [mrad]');
```

```octave
% CASE 2: simulate 1000 turns with a small sexupolar kick each turn
for i=1:1000
  X = R * (X + [ zeros(size(X(1,:))) ; X(1,:) .** 2 ] );
end
```

```octave
% plot
plot(X(1,:), X(2,:), '.');
axis([ -1 1 -1 1 ]);
xlabel('x [mm]');
ylabel('x'' [mrad]');
```
